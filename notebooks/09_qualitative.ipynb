## 9.1 Cascading interdependency matrix

```python
systems = ['Electric Grid', 'Natural Gas Supply', 'Water Supply', 'Telecommunications', 'Transportation']
depends_on = {
    'Electric Grid':        [],
    'Natural Gas Supply':   ['Electric Grid'],
    'Water Supply':         ['Electric Grid'],
    'Telecommunications':   ['Electric Grid'],
    'Transportation':       ['Electric Grid', 'Telecommunications'],
}
n = len(systems)
matrix = np.zeros((n, n))
idx = {s: i for i, s in enumerate(systems)}
for s, deps in depends_on.items():
    for d in deps:
        matrix[idx[s], idx[d]] = 1

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(matrix, cmap='Reds', vmin=0, vmax=1)
ax.set_xticks(range(n)); ax.set_xticklabels(systems, rotation=30, ha='right', fontsize=9)
ax.set_yticks(range(n)); ax.set_yticklabels(systems, fontsize=9)
ax.set_title('Cascading Infrastructure Dependency Matrix\n(1 = row depends on column)')
plt.colorbar(im, ax=ax, shrink=0.7)
plt.tight_layout()
plt.savefig('../data/processed/interdependency_matrix.png', dpi=150)
plt.show()
```
